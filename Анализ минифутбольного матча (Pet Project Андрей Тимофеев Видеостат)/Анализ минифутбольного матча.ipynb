# Цель проекта:

получить количественные и сравнительные характеристики игроков и команд в целом, на основе разбора матча на отдельные действия игроков.

**Описание данных**

Структура файлов разбора 1-го тайма игры:

- игрок_1 — игрок производящий действие;

- игрок_2 - игрок другой команды, при некоторых действиях.

**Действия на поле**

1) Протокольные действия

Протокольные действия (http://www.amfr.ru/ Ассоциация мини-футбола России, ФУТЗАЛ Правила Игры):  

- **Гол** засчитывается, когда мяч полностью пересекает линию ворот между стойками и под перекладиной и при условии, что команда, забившая гол, перед этим не нарушила Правил игры по Футзалу.  

- **Голевой пас** – последний пас забившему игроку, предшествовавший голу. Допускается отсутствие голевого паса при голе. Отскок от отбитого или заблокированного удара не считается голевым пасом.  

- **Фол** - нарушения правил игры наказываются штрафным ударом, пенальти или свободным ударом, фиксируются судьей.  

- **Желтая карточка** – предупреждение судьи за нарушение правил игры.  

- **Красная карточка** – удаление игрока судьей за нарушение правил игры.  

- **Пенальти** – назначается судьей за нарушение игроком в пределах своей штрафной площади.  

- **Угловые** – введение мяча в игру после ухода мяча за лицевую линию от игроков обороняющейся команды.  

- **Аут** - введение мяча в игру после ухода мяча за боковую линию от игроков противоположенной команды.  


2) Игровые действия

Атакующие действия:  

- **Удар (по воротам)** – нацеленный на взятие ворот удар по мячу.  

- **Блокированным удар** считается, если движению мяча помешал полевой игрок обороняющейся команды, засчитывается как удар в створ ворот.  

- **Отбитым ударом** считается, если движению мяча помешал голкипер обороняющейся команды, удар в створ ворот.  

- **Удар мимо (неточный удар)** – если движению мяча не помешали игроки обороняющейся команды, но траектория мяча вне площади ворот. Удары в каркас ворот и удары в игроков атакующей команды засчитываются как удары мимо.  

Пасы  

- **Пас** — нацеленная передача мяча от одного игрока одной команды другому игроку той же команды.  

- **Точный пас** – пас полученный игроком той же команды, пас пришедший в единоборство считается точным.  

- **Острый пас** – умный и своевременный пас, приводящий к значительному улучшению позиций атакующий команды, вывод игрока на голевую позицию.

Контроль мяча, созидающие действия

- **Прием** – обработка мяча после паса, мяч остается под контролем, удобен для следующего действия.  

- **Ведение** – контроль и продвижение мяча к воротам соперника. Формальные признаки: улучшение позиций команды, 3 шага, 3 касания мяча.  

- **Обыгрыш** - уход с мячом после обманного движения, финта, каскада финтов, в свободную, открытую от соперника, зону. Формальные признаки: обманное движение, следующее действие без помехи со стороны соперника, контроль мяча.  

Действия при нейтральном мяче  

- **Подбор** – игрок овладевает мячом без борьбы и активных действий, за счет правильно выбранной позиции на поле, например, после потери контроля мяча соперником или после удара по воротам. Формальные признаки: установление контроля за нейтральным мячом, отсутствие борьбы за мяч.  

- **Единоборство** – борьба за нейтральный мяч с целью сохранить мяч у себя (улучшить позицию, предотвратить угрозу собственных ворот). Успех зависит от физической мощи, быстроты, ловкости, искусности и решительности применения этой силы. Единоборством считается состязание за позицию, используя физический контакт на игровом расстоянии от мяча без применения рук или локтей. Формальные признаки: нейтральный мяч, примирено равные стартовые условия для соперников, участие в борьбе обоих игроков.  

Защитные действия  

- **Перехват** — прерывание точного паса игроком противоположной команды.  

- **Отбор** – активное действие обороняющегося игрока для лишения контроля мяча атакующим игроком. Формальные признаки: активное движение отбирающего, лишение контроля мяча игроком.  

- **Вынос** – безадресное направление мяча при голевой ситуации у собственных ворот.  

- **Заблокировал удар** – помеха движению мяча полевым игроком обороняющейся команды после удара по воротам.

Неудачные действия  

- **Неудачные действия** – действия игрока в простой ситуации, без борьбы со стороны соперника, приведшие к потере мяча, значительному ухудшению позиций команды. Рикошет при голе, неудачный перехват могут быть зафиксированы как неудачное действие.



Составление статистической отчетности:

- **Голы** – количество голов, зафиксированных судьей в протоколе.  

- **Голевые пасы**  – количество голевых пасов, рассчитывается только для игрока.  

- **Фолы** – количество фолов, зафиксированных судьей, без желтых и красных карточек.  

- **Желтая карточка** – количество желтых карточек, зафиксированных в протоколе.  

- **Красная карточка** – количество красных карточек, зафиксированных в протоколе.  

- **Пенальти** – количество проведенных пенальти, зафиксированных в протоколе.  

- **Ввод из аутов и угловых** - количество вводов мяча в игру после выход с площадки.  

- **Отбил ударов ГК** – количество отбитых ударов голкипером, равняется количеству отбитых ударов противоположенной команды.  

- **Удары (по воротам)** – количество ударов по видам ударов. Гол засчитывается как удар в створ.  

- **Точность ударов** – соотношение ударов в створ (отбитые и заблокированные) к общему количеству.  

- **Действий всего** – количество всех зафиксированных игровых действий, состоит из: отбитых голкипером ударов, ведение, все единоборства, все пасы, обыгрыши, отборы, перехваты, подборы, приемы, удары, заблокированных ударов соперника. Неудачные действия не входят в общее количество. Оценка активность игрока.  

- **Пас** — количество пасов по видам пасов.  

- **Точность пасов** – соотношение точных пасов к общему количеству.  

- **Прием** – количество обработок мяча после паса.  

- **Ведение** – количество ведение  мяча к воротам соперника.  

- **Обыгрыш** – количество попыток обыгрышей игроком.  

- **Удачный обыгрыш** – количество удачных обыгрышей игроком.  

- **Удачность обыгрышей** – соотношение удачных обыгрышей к общему количеству.  

- **Подбор** – количество подборов игроком.  

- **Единоборство** – количество всех единоборств в которых участвовал игрок.  

- **Выигранных единоборств** – количество единоборств в которых игрок сохранил мяч у себя (улучшил позицию, предотвратил угрозу собственных ворот).  

- **Выигранных единоборств %** – соотношение выигранных единоборств к общему количеству.  

- **Перехват** — количество удачных перехватов игроком.  

- **Отбор** – количество отборов осуществленных игроком.  

- **Прервал обыгрышей** - количество обыгрышей прерванных игроком.  

- **Вынос** – количество выносов игроком.  

- **Заблокировал ударов** – количество помех движению мяча полевым игроком обороняющейся команды после удара по воротам.  

- **Неудачные действия** – количество неудачных действий игрока.  
- **Был отобран** – количество потерь мяча в результате отбора соперником.

```python
# создадим фукцию для вывода общей инфрмации по таблице целиком
def tab_info(df_x, name):
# получение первых строк таблицы df
    print('Первые строки таблицы', name)
    try:
        display(df_x.head(10))
    except:
        display(df_x.head(2))
# получение случайных строк таблицы df
    print('Cлучайные строки таблицы', name)
    try:
        display(df_x.sample(10, random_state=0))
    except:
        display(df_x.sample(2, random_state=0))
# получение последних строк таблицы df
    print('Последние строки таблицы', name)
    try:
        display(df_x.tail(10))
    except:
        display(df_x.tail(2))
# получение cводной информации по параметрам данных
    print('Сводная информация по параметрам данных', name)
    display(df_x.describe().round(2))
# подсчет количества отсутствующих значений
    print('Количество отсутствующих значений', name)
    display(df_x.isna().sum())
# подсчет доли отсутствующих значений с округлением
    print('Доли отсутствующих значений с округлением', name)
    display(round(df_x.isna().sum() * 100 / len(df_x), 2))
# подсчет количества задублированных записей
    print('Задублированных записей', name)
    display(df_x.duplicated().sum())    
# получение общей информации о данных в таблице
    print('Общая информация о данных в таблице', name)
    display(df_x.info())
```

```python
# создадим функции для контроля меток и цветов

def check_x_left(xx):
    if xx < 12:
        xytext=(-1000, 0)
    else:
        xytext=(-20, 0)
    return (xytext)

def check_x_right(xx):
    if xx < 12:
        xytext=(1000, 0)
    else:
        xytext=(20, 0)
    return (xytext)

def check_color(xx):
    if xx < 50:
        return 'red'
    if xx > 50:
        return 'green'
    else:
        return 'blue'
```

```python
# создадим фукцию для вывода горизонатльной столбчатой диаграммы сравнения долей
# tab_bar_plot(df_tbp, # датесет из 6 столбцов:
# 'Показатели', 'total_1', 'total_2', 'all_total', '%_of_total_1', '%_of_total_2' без n/a
# 'label_left', # подпись левого значения
# 'label_right', # подпись правого значения
# 'title', # название графика
# 'ylabel', # название сравниваемых показателей
# 'color_fon_left', # цвет левого фона
# 'color_text_left', # цвет левого текста
# 'color_fon_right', # цвет правого фона
# 'color_fon_right') # цвет правого текста



def tab_bar_plot(df_tbp, label_left, label_right, title, ylabel,
                   color_fon_left, color_text_left, color_fon_right, color_text_right):
    # переименуем колонки    
    df_tbp.columns = ['Group', 'total_1', 'total_2', 'all_total', '%_of_total_1', '%_of_total_2']
    # размер графика
    fig, ax = plt.subplots(figsize=(16, 10))

    # построение графика по размерам процентных отношений, сумма всегда 100
    ax.barh(df_tbp['Group'], df_tbp['%_of_total_1'], color=color_fon_left, label = label_left)
    ax.barh(df_tbp['Group'], df_tbp['%_of_total_2'], left=df_tbp['%_of_total_1'], color=color_fon_right, label=label_right)
    # подписи значений
    ax.legend(loc='upper center', ncol = 2)
    # название графика и показателей
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    # отступ слева для подписи значений
    ax.set_xlim(xmin = -5)
    # удалим линии графика
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    #ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # зададим количество показателей
    xs = np.arange(len(df_tbp['Group']))
    
    # соберем подписи значений % слева
    ys = df_tbp['%_of_total_1']
    # создадим пары координат
    for x,y in zip(xs,ys):
        # текст подписи
        label = "{:.0f} %".format(y)
        # нанесение подписи по координатам
        plt.annotate(label, # this is the text
                 (y, x), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=check_x_left(y), # distance from text to points (x,y)
                 ha='right', # horizontal alignment can be left, right or center
                 annotation_clip=True, clip_on=True, # настройка видимости 
                 color=color_text_left) # цвет
    # далее по образцу делается для других столбцов
    # также собирается и функция на 2 поля
    ys = df_tbp['total_1']
    zs = df_tbp['%_of_total_1']
    # zip joins x and y coordinates in pairs
    for x,y,z in zip(xs,ys,zs):

        label = "{:.0f}".format(y)

        plt.annotate(label, # this is the text
                 (0, x), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(-5, 0), # distance from text to points (x,y)
                 ha='right', fontsize = 12, color=check_color(z)) # horizontal alignment can be left, right or center


    ys = df_tbp['%_of_total_2']
    # zip joins x and y coordinates in pairs
    for x,y in zip(xs,ys):

        label = "{:.0f} %".format(y)

        plt.annotate(label, # this is the text
                 (100 - y, x), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=check_x_right(y), # distance from text to points (x,y)
                 ha='left',
                 annotation_clip=True, clip_on=True, color=color_text_right) # horizontal alignment can be left, right or center


    ys = df_tbp['total_2']
    zs = df_tbp['%_of_total_2']
    # zip joins x and y coordinates in pairs
    for x,y,z in zip(xs,ys,zs):

        label = "{:.0f}".format(y)

        plt.annotate(label, # this is the text
                 (100, x), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(5, 0), # distance from text to points (x,y)
                 ha='left', fontsize = 12, color=check_color(z)) # horizontal alignment can be left, right or center


    #fig.tight_layout()

    plt.show()
```

```python
# создадим функцию для подсчета ударов команд
def kick_team (df_x):
    df_kick_x = df_x.query('действие in @kick_act')
    df_kick_x = df_kick_x.pivot_table(index='свойство', columns='команда_1', values='время_видео', aggfunc='count')
    df_kick_x = df_kick_x.T.reset_index()
    df_kick_x['точный'] = df_kick_x['гол'] + df_kick_x['заблокированный'] + df_kick_x['отбит']
    df_kick_x = df_kick_x[['точный', 'неточный']].T.reset_index()
    df_kick_x.columns = ['свойство', 'count']
    return df_kick_x
```

```python
# создадим функцию для вывода соотношения и абсолютных значений на графике
def func(pct, allvalues):
    absolute = int(round(pct / 100 * sum(allvalues)))
    return "{:.1f}%\n ({:d})".format(pct, absolute)
# создадим функцию сравнения ударов круговыми диаграмами
def kick_team_contrast (df_x_1, chart_name_1, df_x_2, chart_name_2):
    df_x_left = kick_team(df_x_1)
    df_x_right = kick_team(df_x_2)
    fig, ax = plt.subplots(1, 2, figsize=(16, 4))
    ax[0].pie(df_x_left['count'], labels=df_x_left['свойство'],
              autopct=lambda pct: func(pct, df_x_left['count']), colors = ["g", "r"], 
              explode = [0.1, 0.1], wedgeprops=dict(width=0.8), textprops={'fontsize': 16})
    ax[0].set_title(chart_name_1)
    ax[0].axis("equal")
    ax[1].pie(df_x_right['count'], labels=df_x_right['свойство'], 
              autopct = lambda pct: func(pct, df_x_right['count']),colors = ["g", "r"], 
              explode = [0.1, 0.1], wedgeprops=dict(width=0.8), textprops={'fontsize': 16})
    ax[1].set_title(chart_name_2)
    ax[1].axis("equal")
    plt.show()  
```

```python
# создадим функцию для подсчета пасов команд
def pass_team (df_x):
    df_pass_x = df_x.query('действие in @pass_act')
    df_pass_x = df_pass_x.pivot_table(index='свойство', columns='команда_1', values='время_видео', aggfunc='count')
    df_pass_x = df_pass_x.T.reset_index()
    df_pass_x = df_pass_x[['точный', 'неточный']].T.reset_index()
    df_pass_x.columns = ['свойство', 'count']
    return df_pass_x
```

```python
def func(pct, allvalues):
    absolute = int(round(pct / 100 * sum(allvalues)))
    return "{:.1f}%\n ({:d})".format(pct, absolute)
# создадим функцию сравнения пасов круговыми диаграмами
def pass_team_contrast (df_x_1, chart_name_1, df_x_2, chart_name_2):
    df_x_left = pass_team(df_x_1)
    df_x_right = pass_team(df_x_2)
    fig, ax = plt.subplots(1, 2, figsize=(16, 4))
    ax[0].pie(df_x_left['count'], labels=df_x_left['свойство'],
              autopct=lambda pct: func(pct, df_x_left['count']), colors = ["g", "r"], 
              explode = [0.1, 0.1], wedgeprops=dict(width=0.8), textprops={'fontsize': 16})
    ax[0].set_title(chart_name_1)
    ax[0].axis("equal")
    ax[1].pie(df_x_right['count'], labels=df_x_right['свойство'], 
              autopct = lambda pct: func(pct, df_x_right['count']),colors = ["g", "r"], 
              explode = [0.1, 0.1], wedgeprops=dict(width=0.8), textprops={'fontsize': 16})
    ax[1].set_title(chart_name_2)
    ax[1].axis("equal")
    plt.show()  
```

```python
# создадим функцию для подсчета ударов игроков
def kick_players(df_x, players):
    df_x_kick = df_x.query('действие in @kick_act')
    df_x_kick_all = pd.DataFrame()
    for i in players:
        df_x_kick_player = df_x_kick.query('игрок_1 == @i')
        df_x_kick_player = df_x_kick_player.pivot_table(index='свойство', columns='игрок_1', values='время_видео', aggfunc='count')
        df_x_kick_player = df_x_kick_player.T.reset_index()
        df_x_kick_all = pd.concat([df_x_kick_all, df_x_kick_player])
    df_x_kick_all = df_x_kick_all.fillna(0)
    for column in df_x_kick_all.select_dtypes("float64").columns:
        df_x_kick_all[column] = df_x_kick_all[column].astype('int64')
    df_x_kick_all['точный'] = df_x_kick_all['гол'] + df_x_kick_all['заблокированный'] + df_x_kick_all['отбит']
    df_x_kick_all['точность ударов'] = round((df_x_kick_all['точный'] / (df_x_kick_all['точный'] + df_x_kick_all['неточный']) * 100), 1)
    return(df_x_kick_all.sort_values(by='точность ударов', ascending=False))
```

```python
# создадим функцию для подсчета пасов игроков
def pass_players(df_x, players):
    df_x_pass = df_x.query('действие in @pass_act')
    df_x_pass_all = pd.DataFrame()
    for i in players:
        df_x_pass_player = df_x_pass.query('игрок_1 == @i')
        df_x_pass_player = df_x_pass_player.pivot_table(index='свойство', columns='игрок_1', values='время_видео', aggfunc='count')
        df_x_pass_player = df_x_pass_player.T.reset_index()
        df_x_pass_all = pd.concat([df_x_pass_all, df_x_pass_player])
    df_x_pass_all = df_x_pass_all.fillna(0)
    for column in df_x_pass_all.select_dtypes("float64").columns:
        df_x_pass_all[column] = df_x_pass_all[column].astype('int64')
    df_x_pass_all['точность пасов'] = round((df_x_pass_all['точный'] / (df_x_pass_all['точный'] + df_x_pass_all['неточный']) * 100), 0)
    return(df_x_pass_all.sort_values(by='точность пасов', ascending=False))
```
